```python
def pattern_counter(gene_in, pattern_in):
    i=0
    count=0
    gene=gene_in
    pattern=pattern_in
    while i < len(gene)-len(pattern):
        if(gene[i:len(pattern)+i])==pattern:
            count+=1
        i+=1
    return count
```
